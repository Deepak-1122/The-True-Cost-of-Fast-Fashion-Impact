# Top Brands by Production

```python
top_brands = df.groupby('Brand')['Monthly_Production_Tonnes'].sum().sort_values(ascending=False).head(10)

ax = top_brands.plot(kind='barh', color='skyblue')
plt.title('Top 10 Brands by Monthly Production (Tonnes)')
plt.xlabel('Total Monthly Production')
plt.gca().invert_yaxis()

# Add data labels
for i, v in enumerate(top_brands):
    plt.text(v + 10, i, f'{v:.0f}', va='center')  # 10 is padding

plt.show()
```

# Carbon Emissions by Country

```python
emissions_by_country = df.groupby('Country')['Carbon_Emissions_tCO2e'].sum().sort_values(ascending=False)

fig = px.bar(emissions_by_country.reset_index(), 
             x='Country', y='Carbon_Emissions_tCO2e',
             title='Total Carbon Emissions by Country',
             text='Carbon_Emissions_tCO2e')

fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
fig.show()
```

# Return Rate by Sentiment 

```python
sns.boxplot(data=df, x='Social_Sentiment_Label', y='Return_Rate_Percent', palette='Set2')

# Add mean values as labels
means = df.groupby('Social_Sentiment_Label')['Return_Rate_Percent'].mean()
for i, label in enumerate(means.index):
    plt.text(i, means[label] + 0.5, f'{means[label]:.1f}%', ha='center', color='black')

plt.title('Return Rate by Sentiment')
plt.ylabel('Return Rate (%)')
plt.show()
```

# Water Usage vs Landfill Waste

```python
fig = px.scatter(df, 
                 x='Water_Usage_Million_Litres', 
                 y='Landfill_Waste_Tonnes', 
                 color='Brand',
                 size='Monthly_Production_Tonnes',
                 hover_name='Brand',
                 title='Water Usage vs Landfill Waste by Brand')
fig.show()
```

# Child Labor Incidents

```python
child_labor = df.groupby('Country')['Child_Labor_Incidents'].sum().sort_values(ascending=False).head(10)

ax = child_labor.plot(kind='bar', color='coral')
plt.title('Top 10 Countries by Child Labor Incidents')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45)

# Add data labels
for i, v in enumerate(child_labor):
    plt.text(i, v + 1, str(v), ha='center')

plt.show()
```

# Average Spend vs Sustainability Score

```python
sns.scatterplot(data=df, 
                x='Avg_Spend_Per_Customer_USD', 
                y='Sustainability_Score', 
                hue='Brand')
plt.title('Customer Spend vs Sustainability Score')
plt.xlabel('Average Spend per Customer (USD)')
plt.ylabel('Sustainability Score')
plt.show()
```

# Sentiment vs Social Media Mentions 

```python
df['Sentiment_Size'] = df['Sentiment_Score'].abs() * 20  # Scaled for visibility

fig = px.scatter(df, 
                 x='Instagram_Mentions_Thousands', 
                 y='TikTok_Mentions_Thousands',
                 color='Social_Sentiment_Label',
                 size='Sentiment_Size',
                 hover_name='Brand',
                 title='Social Media Mentions vs Sentiment Score (Bubble Size = |Sentiment| × 20)')

fig.show()
```

# Correlation Heatmap

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

# Carbon Emissions over Years

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Year', y='Carbon_Emissions_tCO2e', hue='Brand', errorbar=None)
plt.title('Carbon Emissions Trend (2015–2024)')
plt.tight_layout()
plt.show()
```

# Social Sentiment Distribution

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=df, x='Social_Sentiment_Label', hue='Brand')
plt.title('Social Media Sentiment by Brand')
plt.tight_layout()

# Add data labels on top of each bar
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.text(p.get_x() + p.get_width() / 2,
                height + 0.2,
                f'{int(height)}',
                ha='center',
                va='bottom',
                fontsize=9,
                color='black',
                fontweight='bold')

plt.show()
```

# Water Usage by Country

```python
plt.figure(figsize=(12, 6))

# Compute average water usage
avg_water = df.groupby('Country')['Water_Usage_Million_Litres'].mean().sort_values()

# Plot and capture the Axes object
ax = sns.barplot(x=avg_water.index, y=avg_water.values, palette='Blues_d')

plt.title('Average Water Usage by Country')
plt.xticks(rotation=45)
plt.ylabel('Million Litres')
plt.tight_layout()

# Add data labels on each bar
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2,
            height + 1,
            f'{height:.1f}',
            ha='center',
            va='bottom',
            fontsize=9,
            color='black',
            fontweight='bold')

plt.show()
```
